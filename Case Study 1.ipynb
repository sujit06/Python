```python
#16.Find:
# a.Mean and std deviation of the income

mean=np.mean(oj.INCOME)
std=np.std(oj.INCOME)
print("Mean of Income=",mean,"Std deviation of the Income=",std)

# b.For income greater than or equal to 10.5, find the mean income

avg=np.mean(oj.INCOME[oj.INCOME>=10.5])
print("Mean of income greater than or equal to 10.5=",avg)

# c.For each brand having price >=2.5 find the mean, median, sd of the log of income

#np.mean(np.log(oj.groupby(["brand"]).INCOME[oj.price>=2.5]))

mean1=oj[oj.price>=2.5].groupby(["brand"]).mean().LogInc
median1=oj[oj.price>=2.5].groupby(["brand"]).median().LogInc
std1=oj[oj.price>=2.5].groupby(["brand"]).std().LogInc
print("Brandwise mean=","\n", mean1,"\n","Brandwise median=","\n",median1,"\n","Brandwise Std Deviation=","\n",std1)
```
