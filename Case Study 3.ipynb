```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ter=pd.read_csv("terror.csv",encoding="latin",low_memory=False)
```

```python
# Q.2)How many attacks happened in India?
sum(ter.country_txt=='India')
```

```python
# Q.2)How many attacks happened in India and upto 3 people were killed?

ter.query("nkill<=3 and country_txt=='India'").shape[0]
```

```python
# Extract the city and summary for attacks above
ter.query("nkill<=3 and country_txt=='India'")[["city","summary"]].dropna(axis=0)
```

```python
# Q.4)In a single terror incident in India, find out top 5 cities by number killed

ter.query("country_txt=='India' and city!='Unknown'").sort_values(by="nkill",ascending=False).head(5).city


```

```python
#Q.5)In a single terror incident in India, find out top 5 cities by number killed and wounded 
ter2=ter.query("country_txt=='India'")
ter2['total']=ter2.nkill+ter2.nwound
ter2.sort_values(by='total',ascending=False).head(5).city
```

```python
#How many attacks were successful that were suicide attacks?
ter.query("suicide==1 and success==1").shape[0]
```

```python
#label all the incidents where the number killed was more than 5 as severe. 
ter["label"]=""
ter.label[ter.nkill>5]="severe"
```

```python
#) write a function to label an incident that was both successful and suicidal
def func(ter):
    ter.label[np.logical_and(ter.success==1,ter.suicide==1)]="suicide_success"
func(ter)
```

```python
#Q.9)Create a new category representing if the incident occured in Afghanistan, Pakistan or India 
#as one level of the category and all the other countries as another level
ter["category"]=""
ter.category[(ter.country_txt=='India') | (ter.country_txt=='Afghanistan') | (ter.country_txt=='Pakistan')]="Af-Pak-India"
ter.category[(ter.country_txt!='India') & (ter.country_txt!='Afghanistan') & (ter.country_txt!='Pakistan')]="ROW"
```

```python
#Q.10)How many incidents happened in Af-Pak-India vs ROW?
print("Af-Pak-India= ",sum(ter.category=="Af-Pak-India"), "ROW= ",sum(ter.category=="ROW"))
```

```python
#Q.11)List the number of suicides attacks and average kills by Af-Pak-India vs ROW.
# Rename columns in the output as Average_Kills and Number_Incidents

ter.groupby("category").agg({'suicide':np.sum,'nkill':np.mean})
```
